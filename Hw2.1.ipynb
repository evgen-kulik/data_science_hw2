#### 1.1. Вивести перші рядки таблиці з посилання за допомогою методу head

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
list_of_tables = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", match="Регіон")
df = list_of_tables[0]
print(df.head())
```

#### 1.2. Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
print(df.shape)
```

#### 1.3. Замініть у таблиці значення "—" на значення NaN

```python
df = df.replace('—', 'NaN')
print(df)
```

#### 1.4. Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
print(df.dtypes)
```

#### 1.5. Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
# оскільки в колонках присутнє NaN, то в int64 перевести тип не маємо можливості, тому використовуємі float64
df['2014'] = df['2014'].astype('float64')
df['2019'] = df['2019'].astype('float64')
print(df.dtypes)
```

#### 1.6. Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
# проаналізуємо весь df на предмет відсутніх даних в комірках
nan_value = df.isnull()
print(nan_value)
```

```python
# підрахуємо кількість порожніх комірок по всьому df
count_nan = nan_value.sum()
print(count_nan)
```

#### 1.7. Видаліть з таблиці дані по всій країні (останній рядок таблиці)

```python
df = df.drop(27)
print(df)
```

#### 1.8. Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df[['1950', '1960', '1970', '2014', '2019']] = df[['1950', '1960', '1970', '2014', '2019']]. fillna (df[['1950', '1960', '1970', '2014', '2019']].mean())
print(df)
```

#### 1.9. Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
# визначемо середнє значення народжуваності у 2019 році
avg_2019 = df['2019'].mean()
print(avg_2019)
regions_list = df.loc[df['2019'] > avg_2019]['Регіон']
print(regions_list)
```

#### 1.10. У якому регіоні була найвища народжуваність у 2014 році?

```python
region =  df[df['2014']==df['2014'].max()]['Регіон']
print(region)
```

#### 1.11. Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
fig, ax = plt.subplots(figsize=(8,8))
plt.barh(df['Регіон'], df['2019'], color="g")
plt.xlabel("Коефіцієнт народжуваності", fontsize="small", color="g")
plt.ylabel("Регіон", fontsize="small", color="g")
plt.title("Коефіцієнт народжуваності у регіонах України у 2019", fontsize=15, color="b")
plt.show()
```

#### 1.12. Візуалізація. Додайте від 3 до 5 графіків функцій різного типу на ваш вибір. Задайте графікам оформлення, щоб кожен графік чимось відрізнявся і не був схожим на інші. Можна використовувати як matplotlib, так і seaborn

```python
# 1й графік
fig, ax = plt.subplots(figsize=(8,8))
plt.plot(df['Регіон'], df['2019'], color="r")
plt.xticks(rotation=90)
plt.xlabel("Коефіцієнт народжуваності", fontsize="small", color="r")
plt.ylabel("Регіон", fontsize="small", color="r")
plt.title("Коефіцієнт народжуваності у регіонах України у 2019", fontsize=15, color="r")
plt.show()
```

```python
# 2й графік
fig, ax = plt.subplots(figsize=(8,8))
plt.plot(df['Регіон'], df['2019'], color="r", label="2019")
plt.plot(df['Регіон'], df['2014'], color="b", label="2014")
plt.xticks(rotation=90)
plt.xlabel("Коефіцієнт народжуваності", fontsize="small", color="r")
plt.ylabel("Регіон", fontsize="small", color="r")
plt.title("Коефіцієнт народжуваності у регіонах України у 2014 та 2019", fontsize=15, color="r")
plt.legend()
plt.show()
```

```python
# 3й графік
fig, ax = plt.subplots(figsize=(8,8))
plt.plot(df['Регіон'], df['2019'], color="r", linestyle="--", linewidth=2, marker="o", label="2019")
plt.plot(df['Регіон'], df['2014'], color="b", linestyle="-", linewidth=2, marker="D", label="2014")
plt.plot(df['Регіон'], df['2000'], color="g", linestyle=":", linewidth=2, marker="v", label="2000")
plt.xticks(rotation=90)
plt.xlabel("Коефіцієнт народжуваності", fontsize="small", color="r")
plt.ylabel("Регіон", fontsize="small", color="r")
plt.title("Коефіцієнт народжуваності у регіонах України у 2000, 2014 та 2019", fontsize=15, color="r")
plt.legend()
plt.show()
```

```python
# 4й графік
labels = ['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']
data = [df.at[0, '1950'], df.at[0, '1960'], df.at[0, '1970'], df.at[0, '1990'], df.at[0, '2000'], df.at[0, '2012'], df.at[0, '2014'], df.at[0, '2019']]
explode = [0.15, 0, 0, 0, 0.15, 0, 0, 0]
plt.pie(
    data,
    labels=labels,
    shadow=True,
    explode=explode,
    autopct="%.2f%%",
    pctdistance=0.7,
    labeldistance=1.1,
)
plt.title("Розподіл коефіцієнта народжуваності в Криму в різні роки", fontsize=15, color="b", pad=15)
plt.show()
```

```python
# 5й графік
labels = ['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']
data = [df.at[0, '1950'], df.at[0, '1960'], df.at[0, '1970'], df.at[0, '1990'], df.at[0, '2000'], df.at[0, '2012'], df.at[0, '2014'], df.at[0, '2019']]

plt.pie(
    data,
    labels=labels,
    wedgeprops=dict(width=0.5),
    autopct="%.2f%%",
    pctdistance=0.8,
    labeldistance=1.1
)
plt.title("Розподіл коефіцієнта народжуваності в Криму в різні роки", fontsize=15, color="y", pad=15)
plt.show()
```
